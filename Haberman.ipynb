**Classification Haberman using Decision Tree**

```
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```
data_column = dataset[['Age of patient at time of operation', 'Patients year of operation', 'Number of positive axillary nodes detected']].values

data_class = dataset[['class']].values
```

```
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(data_column, data_class, train_size=0.7, random_state=1)
```

```
dtc = DecisionTreeClassifier()
```

```
dtc.fit(train_inputs, train_classes)
```

**KEAKURATAN**

```
dtc.score(test_inputs, test_classes)
```

```
y_pred = dtc.predict(test_inputs)
```

```
from sklearn.metrics import classification_report
print(classification_report(test_classes, y_pred))
```
